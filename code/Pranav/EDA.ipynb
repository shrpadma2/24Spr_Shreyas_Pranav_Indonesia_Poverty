```python
import pandas as pd
from dbfread import DBF
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
pd.set_option('display.max_rows', 500)
```

```python
def readdbf(dbfpath):
    table = DBF(dbfpath, load=True)
    df = pd.DataFrame(iter(table))
    return df
```

## Understanding 2020 

```python
df1 = readdbf('../data/Annual Area-based Poverty Map/PODES/2020/podes2020_b1b7.dbf')
print(df1.shape)
df1.head()
```

```python
df2 = readdbf('../data/Annual Area-based Poverty Map/PODES/2020/podes2020_b8b13.dbf')
print(df2.shape)
df2.head()
```

```python
print(df1.isnull().sum())
print(df2.isnull().sum())
```

```python
df = pd.merge(df1, df2)
df.head()
```

```python
df.shape
```

## Understanding Electric usage

PLN: Perusahaan Listrik Negara (State Electricity Company) is an Indonesian government-owned corporation which has a monopoly on electricity distribution in Indonesia and generates the majority of the country's electrical power.

### Working with columns ("R501A1", "R501A2", "R501B", "R502A", "R502B")

* R501A1	Number of families of PLN electricity users
* R501A2	Number of Non PLN Electric User Families
* R501B	Number of Families Non -Electric User
* R502A	Lighting on the main road/kelurahan road
* R502B	If there is lighting on the main road of the village/kelurahan (R402A coded 1 or 2), the type of lighting


```python
electric_cols = ["R101N", "R102N", "R103N", "R104N", "R501A1", "R501A2", "R501B", "R502A", "R502B"]
electric_df = df[electric_cols]
electric_df.head()
```

```python
electric_df['R501A2'].value_counts()
```

```python
province_electricity = electric_df.groupby("R101N")["R501A1"].sum()
province_electricity = province_electricity.sort_values(ascending=False)
```

### Province wise number of families of PLN electricity users

```python
plt.figure(figsize=(20, 10))
sns.barplot(x=province_electricity.index, y=province_electricity.values)
plt.xticks(rotation=90)
```

```python
province_electricity_nonpla = electric_df.groupby("R101N")["R501A2"].sum()
province_electricity_nonpla = province_electricity_nonpla.sort_values(ascending=False)

plt.figure(figsize=(20, 10))
sns.barplot(x=province_electricity_nonpla.index, y=province_electricity_nonpla.values)
plt.xticks(rotation=90)
```

```python
grouped_data = electric_df.groupby('R101N')[['R501A1', 'R501A2']].sum()

# Sort the values by R501A1 (You could also sort by the total if preferred)
grouped_data = grouped_data.sort_values(by='R501A1', ascending=False)

# Create a stacked bar plot
grouped_data.plot(kind='bar', stacked=True, figsize=(20, 10))

# Setting the plot titles and labels
plt.title('Electricity Users by Province')
plt.xlabel('Province')
plt.ylabel('Number of Users')
plt.xticks(rotation=90)  # Rotate the x-axis labels for better readability

plt.legend(['PLN Electric Users', 'Non-PLN Electric Users'])
plt.show()
```

```python
total_counts = electric_df.groupby('R101N')['R502A'].count()

# Then, calculate the counts for each category within 'R502A' per 'R101N'
category_counts = electric_df.groupby(['R101N', 'R502A'])['R502A'].count()

# Calculate the percentage
percentage = (category_counts / total_counts.reindex(category_counts.index, level='R101N')) * 100

# Convert to DataFrame for better readability
percentage_df = percentage.reset_index(name='Percentage')

percentage_df
```

```python
# Renaming the R502A values
temp_df = percentage_df.copy()

temp_df['R502A'] = temp_df['R502A'].map({1: "Yes-mostly", 2: "Yes-small part", 3: "None"})

# Pivot the DataFrame to get it ready for stacked bar plot
df_pivot = temp_df.pivot(index='R101N', columns='R502A', values='Percentage')
df_pivot_sorted = df_pivot.sort_values(by="None", ascending=False)


# Plotting
plt.figure(figsize=(30, 7))

df_pivot_sorted.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.ylabel('Percentage')
plt.title('Percentage distribution of lighting on roads by Province')
plt.xticks(rotation=45)
plt.legend(title='lighting on roads', loc='upper left')
plt.xticks(rotation=90)

plt.show()
```

## water sources

```python
water_cols = ["R101N", "R102N", "R103N", "R104N", "R508", "R509"]
water_df = df[water_cols]
water_df.head()
```

```python
r508 = {
    1: "Branded bottled water", 
2: "Well", 
3: "Metered plumbing", 
4: "Unmetered plumbing"
, 5: "Drilled well or pump"
, 6: "Refill water"
, 7: "Springs"
, 8: "River/lake/pond"
, 9: "Rainwater"
, 10: "Others"
}

r509 = {
    
    1: "Metered plumbing (PAM/PDAM)"
, 2: "Unmetered plumbing"
, 3: "Drilled well or pump"
, 4: "Rainwater"
, 5: "Spring"
, 6: "River/lake/pond"
, 7: "Wells"
}

water_df['R508'] = water_df['R508'].map(r508)
water_df['R509'] = water_df['R509'].map(r509)

water_df.head()
```

```python
plt.figure(figsize=(20, 10))
sns.barplot(x = water_df['R508'].value_counts().index, y = water_df['R508'].value_counts().values)
plt.title('Drinking Water Source Distribution')
```

```python
plt.figure(figsize=(20, 10))
sns.barplot(x = water_df['R509'].value_counts().index, y = water_df['R509'].value_counts().values)
plt.title('Bathing Water Source Distribution')
```
