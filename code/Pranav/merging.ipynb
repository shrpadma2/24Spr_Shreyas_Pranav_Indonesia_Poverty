```python
import pandas as pd
from dbfread import DBF
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
pd.set_option('display.max_rows', 500)
```

```python
def readdbf(dbfpath):
    table = DBF(dbfpath, load=True)
    df = pd.DataFrame(iter(table))
    return df
```

```python
df1 = readdbf('../data/Annual Area-based Poverty Map/PODES/2020/podes2020_b1b7.dbf')

df2 = readdbf('../data/Annual Area-based Poverty Map/PODES/2020/podes2020_b8b13.dbf')

df = pd.merge(df1, df2)

```

```python
df["Primary_col"] = df['R101'] + df["R102"]
df.head()
```

```python
df.iloc[:,-15:].head()
```

## Aggregating the data

```python
agg_df = pd.read_csv("../working folder/aggregation_method_1.csv")
agg_df.head()
```

```python
agg_df['aggregate by '].unique()
```

```python
dictionary_zip = dict(zip(agg_df['Kode Variabel'], agg_df['aggregate by '].replace('nan', None)))
```

```python
# Filter out the NaN values which represent columns not to be aggregated
agg_dict = {k: v for k, v in dictionary_zip.items() if pd.notna(v)}

def mode_agg(series):
    modes = series.mode()
    return modes[0] if not modes.empty else None

def proportion_agg(series):
    value_counts = series.value_counts(normalize=True)
    return value_counts.iloc[0] if not value_counts.empty else np.nan
```

```python
agg_funcs = {
    'mode': mode_agg,
    'sum': 'sum',  # 'sum' can be used directly
    'proportion': proportion_agg,
    'mean' : "mean"
}
agg_dict = {k: agg_funcs[v] for k, v in agg_dict.items()}


aggregated_df = df.groupby('Primary_col').agg(agg_dict)
```

## 34 BPS codes have null values

```python
crosswalk.dropna(subset=['bps_code'], inplace=True)
crosswalk['bps_code'] = crosswalk['bps_code'].astype(int).astype(str)
crosswalk.head()
```

```python
## BPS code and primary col comparision

full_list = aggregated_df['Primary_col'].unique()
crosswalk_bps = crosswalk['bps_code'].unique()
missing_numbers = sorted(set(full_list) - set(crosswalk_bps))
missing_numbers
```

```python
podus_crosswalk = aggregated_df.merge(crosswalk, left_on='Primary_col', right_on='bps_code', how='left')
podus_crosswalk.head()
```
