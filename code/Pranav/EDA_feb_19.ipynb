```python
import pandas as pd
from dbfread import DBF
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
pd.set_option('display.max_rows', 500)
```

```python
def readdbf(dbfpath):
    table = DBF(dbfpath, load=True)
    df = pd.DataFrame(iter(table))
    return df
```

```python
df['PLN users by 1000'] = df["R501A1"]/df['Total Families']*1000
df['Non-PLN users by 1000'] = (df["R501A2"] + df["R501B"])/df['Total Families']*1000
df['Non electric users by 1000'] = (df["R501B"])/df['Total Families']*1000
df.head()
```

```python
grouped_data = df.groupby('R101N')[["PLN users by 1000", "Non-PLN users by 1000"]].mean()

# Sort the values by R501A1 (You could also sort by the total if preferred)
grouped_data = grouped_data.sort_values(by='PLN users by 1000', ascending=False)

# Create a stacked bar plot
grouped_data.plot(kind='bar', stacked=True, figsize=(20, 10))

# Setting the plot titles and labels
plt.title('Electricity Users by Province for 1000 Families')
plt.xlabel('Province')
plt.ylabel('Number of Users')
plt.xticks(rotation=90)  # Rotate the x-axis labels for better readability

plt.legend(['PLN Electric Users', 'Non-PLN Electric Users'])
plt.show()
```

```python
percentage_df = df.groupby('R101N')[['Total Families']].sum()
percentage_df[['Total Families']].sum()
```

```python
total_counts = df.groupby('R101N')['R502A'].count()

# Then, calculate the counts for each category within 'R502A' per 'R101N'
category_counts = df.groupby(['R101N', 'R502A'])['R502A'].count()

# Calculate the percentage
percentage = (category_counts / total_counts.reindex(category_counts.index, level='R101N')) * 100

# Convert to DataFrame for better readability
percentage_df = percentage.reset_index(name='Percentage')


percentage_df['R502A'] = percentage_df['R502A'].map({1: "Yes-mostly", 2: "Yes-small part", 3: "None"})

# Pivot the DataFrame to get it ready for stacked bar plot
df_pivot = percentage_df.pivot(index='R101N', columns='R502A', values='Percentage')
df_pivot_sorted = df_pivot.sort_values(by="None", ascending=False)

# Plotting with specified colors
colors = {'Yes-mostly': 'tab:green', 'Yes-small part': 'tab:blue', 'None': 'tab:orange'}

# Order the colors in the same order as the columns
ordered_colors = [colors[column] for column in df_pivot_sorted.columns]


# Plotting
plt.figure(figsize=(30, 7))

df_pivot_sorted.plot(kind='bar', stacked=True,color=ordered_colors, figsize=(10, 6))
plt.ylabel('Percentage')
plt.title('Percentage distribution of lighting on roads by Province')
plt.xticks(rotation=45)
plt.legend(title='lighting on roads', bbox_to_anchor=(1.0, 1.0))
plt.xticks(rotation=90)

plt.show()
```

```python
r508 = {
    1: "Branded bottled water", 
2: "Well", 
3: "Metered plumbing", 
4: "Unmetered plumbing"
, 5: "Drilled well or pump"
, 6: "Refill water"
, 7: "Springs"
, 8: "River/lake/pond"
, 9: "Rainwater"
, 10: "Others"
}

temp_df = df.copy()
temp_df["R508"] = temp_df["R508"].map(r508)
temp_df["R508"].value_counts().index
```

```python
unsafe = ['Metered plumbing', 'Drilled well or pump', 'Refill water', 'Branded bottled water']
unsafe = ['River/lake/pond', 'Unmetered plumbing', 'Springs', 'Well', 'Rainwater', 'Others']

# Function to categorize water source
def categorize_water_source(source):
    if source in unsafe:
        return 'safe'
    elif source in unsafe:
        return 'unsafe'
    else:
        return 'Unknown'

# Apply the function to create a new column
temp_df['Category'] = temp_df['R508'].apply(categorize_water_source)

temp_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches
import textwrap

# Assuming temp_df is your DataFrame and it has a 'Category' column
# plt.figure(figsize=(20, 10))
vc = temp_df['Category'].value_counts()
vc = vc.sort_values(ascending=True)
sns.barplot(x=vc.index, y=vc.values, hue=vc.index)
plt.title('Drinking Water Source Distribution')
plt.xlabel('Drinking water safety')

# Lists of safe and unsafe sources
safe_sources = ['Metered plumbing', 'Drilled well or pump', 'Refill water', 'Branded bottled water']
unsafe_sources = ['River/lake/pond', 'Unmetered plumbing', 'Springs', 'Well', 'Rainwater', 'Others']

# Function to wrap text for each source category
def wrap_text(source_list):
    return '\n'.join(textwrap.wrap(', '.join(source_list), width=30))

# Create custom patches with wrapped text
safe_patch = mpatches.Patch(color=sns.color_palette()[0], label='Safe Sources:\n' + wrap_text(safe_sources))
unsafe_patch = mpatches.Patch(color=sns.color_palette()[1], label='Unsafe Sources:\n' + wrap_text(unsafe_sources))

# Place the custom legend on the plot, using 'best' location to automatically adjust
plt.legend(handles=[safe_patch, unsafe_patch], title='Source Categories',  bbox_to_anchor=(1.0, 1.0), fontsize='small')


plt.tight_layout()
plt.show()
```

```python
total_families = df.groupby('R101N')[['Total Families']].sum().reset_index()
```

```python
x = df.groupby('R101N')['R512B3'].sum().reset_index().merge(total_families, on='R101N')
x['number of families in slums for 1000'] = (x['R512B3']/x['Total Families']*1000).astype(int)
x.sort_values(by='number of families in slums for 1000', ascending=False, inplace=True)
x
```

```python

# Plotting
fig, ax1 = plt.subplots(figsize=(14, 8))

# Bar plot for Total Families
sns.barplot(x='R101N', y='Total Families', data=x, ax=ax1, color='lightblue', label='Total Families')
ax1.set_xlabel('Province')
ax1.set_ylabel('Total Families', color='b')
ax1.tick_params(axis='y', labelcolor='b')
ax1.set_xticklabels(x['R101N'], rotation=90)

# Create a second y-axis for Families in Slums per 1000
ax2 = ax1.twinx()
sns.lineplot(x='R101N', y='number of families in slums for 1000', data=x, ax=ax2, color='darkorange', marker='o', label='Families in Slums per 1000 families')
ax2.set_ylabel('Families in Slums per 1000 families', color='darkorange')
ax2.tick_params(axis='y', labelcolor='darkorange')

# Adding legend manually
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.title('Number of Families living in slums for 1000 Families by Province')
plt.tight_layout()
plt.show()
```

### If we had number of beds in a hospital and number of people in the families, we would get much accurate results of number of beds available for 1000 people.(which is a standard measure of hospital beds)

```python
hospital_df = df.groupby('R101N').agg({'R603AK2':'sum', "R603AK3":'mean'}).reset_index().merge(total_families, on='R101N')
hospital_df['number of hospitals for 100000 families'] = (hospital_df['R603AK2']/hospital_df['Total Families']*100000)
hospital_df.sort_values(by='R603AK3', ascending=False, inplace=True)
national_average_for_100000 = hospital_df['number of hospitals for 100000 families'].mean()
```

```python
plt.figure(figsize=(14, 8))
ax1 = sns.barplot(x='R101N', y='number of hospitals for 100000 families', data=hospital_df, color='lightblue', label='Number of Hospitals per 100000 Families')

# Set the x-axis labels and rotate them for better readability
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45, ha="right")

# Create a secondary axis
ax2 = ax1.twinx()

# Plot the average distance to the nearest hospital on the secondary axis
sns.lineplot(x='R101N', y='R603AK3', data=hospital_df, ax=ax2, color='darkorange', marker='o', label='Average Distance to Nearest Hospital')
ax1.axhline(national_average_for_100000, color='red', linestyle='--', label='National Average: 4.5')



# Labeling the axes
ax1.set_xlabel('Province')
ax1.set_ylabel('Number of Hospitals per 100000 Families', color='blue')
ax2.set_ylabel('Average Distance to Nearest Hospital', color='darkorange')

# Adding legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.title('Number of Hospitals per 100000 Families and Average Distance to Nearest Hospital by Province')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

data_dict = {   "R101N": "Province name",
                "R605AK4": "People died of vomiting/diarrhea", 
                "R605BK4": "People died of dengue fever", 
                "R605CK4":	"People died of measles", 
                "R605DK4":	"People died of malaria", 
                "R605EK4":	"People died of bird flu/sars", 
                "R605FK4":	"People died of hepatitis e", 
                "R605GK4":	"People died of diphtheria DPT", 
                "R605HK4":	"People died of Covid-19", 
                "R605IK4":	"People died of other (chikungunya, leptospirosis, cholera, etc.)"
            }

death_df = df.copy()
death_df = death_df[["R101N","Total Families" , "R605AK4", "R605BK4", "R605CK4", "R605DK4", "R605EK4", "R605FK4", "R605GK4", "R605HK4", "R605IK4"]]

death_df.rename(columns=data_dict, inplace=True)

agg_df = death_df.groupby("Province name").sum().reset_index()

for column in agg_df.columns[2:]:
    agg_df[column] = (agg_df[column] / agg_df['Total Families']) * 1000

def create_interactive_plot_adjusted(df):
    diseases = agg_df.columns[2:]  
    fig_width = 1100
    fig_height = 700
    
    fig = make_subplots(specs=[[{"secondary_y": True}]])

    fig.add_trace(go.Bar(x=agg_df['Province name'], y=agg_df['Total Families'], name='Total Families'), secondary_y=False)

    for disease in diseases:
        fig.add_trace(go.Scatter(x=agg_df['Province name'], y=agg_df[disease], name=disease, visible="legendonly"), secondary_y=True)

    fig.update_layout(
        title_text="Number of Deaths per 1000 Families by Province",
        width=fig_width, 
        height=fig_height,  
        legend_title_text='Disease',
        legend=dict(
            orientation="h",  
            yanchor="bottom",
            y=-0.7,  
            xanchor="center",
            x=0.5
        ),
        margin=dict(l=0, r=0, t=100, b=200)  
    )

    # Set axis titles
    fig.update_xaxes(title_text="Province", tickangle=-45)  
    fig.update_yaxes(title_text="Total Families in province", secondary_y=False)
    fig.update_yaxes(title_text="Number of Deaths per 1000 Families", secondary_y=True)

    fig.update_layout(
        font=dict(
            size=12,  
        )
    )

    fig.show()

create_interactive_plot_adjusted(agg_df)
```
